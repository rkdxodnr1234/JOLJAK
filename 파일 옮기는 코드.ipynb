```python
import os
import shutil

# 이미지 파일이 포함된 최상위 폴더 경로
source_folders = [
    "C:/Users/ahsld/Downloads/098.지능형 스마트팜 통합 데이터(파프리카)/01.데이터/1.Training",

]

# 파일을 이동할 목적지 경로 (모든 파일을 이 경로로 이동)
destination_folder = "C:/Users/ahsld/Desktop/PAPRIKA/PAP_Train"

# 파일 이동 함수
def move_files_to_root(source_folders, destination_folder):
    # 각 폴더를 순회
    for folder in source_folders:
        # 하위 폴더 내 모든 파일을 순회
        for root, dirs, files in os.walk(folder):
            for file in files:
                # .png 파일과 .txt 파일만 이동
                if file.lower().endswith(('.png', '.txt')):
                    # 원본 파일 경로
                    source_path = os.path.join(root, file)
                    
                    # 목적지 파일 경로
                    destination_path = os.path.join(destination_folder, file)
                    
                    try:
                        # 파일 이동
                        shutil.move(source_path, destination_path)
                        print(f"Moved: {source_path} to {destination_path}")
                    except shutil.Error as e:
                        print(f"Error moving {source_path}: {e}")

# 실행
move_files_to_root(source_folders, destination_folder)
```

txt파일만 옮기기

```python
import os
import shutil

# 원본 폴더 경로와 대상 폴더 경로를 설정합니다.
source_dir = "F:/Training/image"  # 원본 폴더 경로
destination_dir = "D:/이매지니어/image"  # 복사할 대상 폴더 경로

# 대상 폴더가 존재하지 않으면 생성합니다.
os.makedirs(destination_dir, exist_ok=True)

# txt 파일만 복사
file_count = 0
skipped_count = 0
for root, dirs, files in os.walk(source_dir):
    for file in files:
        if file.endswith('.txt'):
            src_path = os.path.join(root, file)
            dst_path = os.path.join(destination_dir, file)
            
            # 동일한 이름의 파일이 대상 폴더에 존재하는지 확인
            if os.path.exists(dst_path):
                skipped_count += 1
                print(f"Skipped: {file} (already exists)")
            else:
                shutil.copy2(src_path, dst_path)
                file_count += 1
                # 100개마다 진행 상황을 출력합니다.
                if file_count % 100 == 0:
                    print(f"{file_count} files copied...")

print(f"Total {file_count} .txt files copied to {destination_dir}")
print(f"Total {skipped_count} files skipped (already existed)")
```
